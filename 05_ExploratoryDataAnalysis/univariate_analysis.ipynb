# Univariate Analysis

Univariate analysis involves the examination and analysis of each variable in a dataset independently. It's used to describe patterns and find insights within single variables, as opposed to bivariate or multivariate analysis, which involves two or more variables. Univariate analysis can include various statistical and visualization techniques depending on the type of the variable (categorical or numerical). ## 1. Categorical Data

### a. Countplot

```python
import seaborn as sns
import pandas as pd
```

```python
pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv("../datasets/titanic.csv")
print(f"shape: {df.shape}")
print("=" * 150)
print(df.sample(5))
```

```python
print(df['survived'].value_counts())
sns.countplot(data=df, x='survived', hue='survived')
```

```python
sns.countplot(data=df, x='pclass', hue='pclass')
```

```python
sns.countplot(data=df, x='sex', hue='sex')
```

```python
sns.countplot(data=df, x='embarked', hue='embarked')
```

### b. Pie chart

```python
import matplotlib.pyplot as plt
```

```python
col = 'survived'
survived_category_counts = df[col].value_counts()
print(f"survived category counts: {survived_category_counts}")
plt.pie(x=survived_category_counts, labels=survived_category_counts.index, autopct='%.2f%%')
plt.title(col, loc='left')
plt.show()
```
