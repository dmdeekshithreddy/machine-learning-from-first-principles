Taken a dataset to explain how linear regression algorithm is written and model learns if the input feature has negative values

```python
# imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the input data
input_df = pd.read_csv('datasets/energy_used.csv')
input_df.head()
```

```python
# show the data in a scatter plot with quadrants and labels, with origin and horizontal line at the center
plt.figure(figsize=(8, 6))
plt.scatter(input_df['Outside_Temp_C'], input_df['Energy_Consumption_kWh'])
plt.xlabel('Outside Temperature in Celsius')
plt.ylabel('Energy Used')
plt.title('Energy Used vs Outside Temperature with Quadrants')

# Calculate center for horizontal and vertical lines
x_center = (input_df['Outside_Temp_C'].max() + input_df['Outside_Temp_C'].min()) / 2
y_center = (input_df['Energy_Consumption_kWh'].max() + input_df['Energy_Consumption_kWh'].min()) / 2

# Draw the center lines
plt.axhline(y_center, color='black', linewidth=1)
plt.axvline(x_center, color='black', linewidth=1)

# Get axis limits for placing quadrant labels
xlim = plt.xlim()
ylim = plt.ylim()

# Place quadrant labels
plt.text((xlim[1]+x_center)/2, (ylim[1]+y_center)/2, 'Q1', fontsize=14, color='red', ha='center', va='center')  # +x, +y
plt.text((xlim[0]+x_center)/2, (ylim[1]+y_center)/2, 'Q2', fontsize=14, color='blue', ha='center', va='center') # -x, +y
plt.text((xlim[0]+x_center)/2, (ylim[0]+y_center)/2, 'Q3', fontsize=14, color='green', ha='center', va='center') # -x, -y
plt.text((xlim[1]+x_center)/2, (ylim[0]+y_center)/2, 'Q4', fontsize=14, color='purple', ha='center', va='center') # +x, -y

# Mark the origin (center)
plt.scatter([x_center], [y_center], color='black', s=50, zorder=5)
plt.text(x_center, y_center, 'Origin', fontsize=12, color='black', ha='left', va='bottom')

plt.show()
```
