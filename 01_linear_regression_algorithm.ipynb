```python
def square_trick(learning_rate, price_per_room, base_price, number_of_rooms, actual_price):
    predicted_price = price_per_room * number_of_rooms + base_price
    error = actual_price - predicted_price
    new_base_price = base_price + (learning_rate * error)
    new_price_per_room = price_per_room + (learning_rate * number_of_rooms * error)
    return new_price_per_room, new_base_price
```

<br> Step1: start with random `base_price` and `price_per_room` between 0 and 1 ; 
<br> Step2: use random.random() function from Python’s built-in random module that returns a random floating-point number between 0.0 (inclusive) and 1.0 (exclusive).
<br> Step3: for each epoch pick a random sample, calculate the price with current slope (price_per_room) and y-intercept (base_price)
<br> Step4: compare with actual price to calculate the error and adjust the base_price and price_per_room as per the error

```python
import random
def linear_regression(features, labels, learning_rate=0.01, epochs = 1000): 
    base_price = random.random()
    price_per_room = random.random()
    for epoch in range(epochs):
        i = random.randint(0, len(features) - 1)
        number_of_rooms = features[i]
        actual_price = labels[i]
        price_per_room, base_price = square_trick(learning_rate, price_per_room, base_price, number_of_rooms, actual_price)

    return price_per_room, base_price
```
